## $\tau(R)$

```python
def Sigma_g(R, Sigma0, gama, R_in, R_out):

	Sigma = Sigma0 * (R/R_out)**(-gama) \
			* np.exp(-(R/R_out)**(2-gama)) \
			* np.exp(-(R/R_in)**(gama-2)) 
	
	return Sigma


def Temp_from_R(path, N):

	with open(f"{path}/r_out-{N:04d}.dat") as input_file:
		time = [next(input_file) for _ in range(2)][1].split()[1]
		input_file.readline()
		input_file.readline()
		head = input_file.readline().split()[1:]
	
	df = pd.read_table(f"{path}/r_out-{N:04d}.dat", sep=r"\s+", comment='#', names=head)

	df = df[df.j == 100]
	df.drop(df.tail(1).index, inplace=True) # drop last 1 rows

	T = np.array(df["T"])
	R = np.array(df["R_c"]/AU)

	return T, R


def tau(path, N, R, T, f_d, Sigma0, gama, R_in, R_out):

	df_kappa = pd.read_table(f"{path}/PRmix_{N}.txt", skiprows=1, sep="\s+")
	y_interp_P = scipy.interpolate.interp1d(df_kappa["temperature[K]"], df_kappa["kappa_P[cm2/g]"]/f_d)
	y_interp_R = scipy.interpolate.interp1d(df_kappa["temperature[K]"], df_kappa["kappa_R[cm2/g]"]/f_d)

	kappaP_arr = np.empty(len(T))
	kappaR_arr = np.empty(len(T))

	for i, t in enumerate(T):
		kappaP_arr[i] = float(y_interp_P(t))
		kappaR_arr[i] = float(y_interp_R(t))


	Sigma_gas = Sigma_g(R, Sigma0, gama, R_in, R_out) / 2
	
	tauP = f_d * Sigma_gas * kappaP_arr
	tauR = f_d * Sigma_gas * kappaR_arr

	return tauP, tauR
```

```python
def Tmax_fun(file_rout_max):

	with open(f"{file_rout_max}") as input_file:
		time = [next(input_file) for _ in range(2)][1].split()[1]
		input_file.readline()
		input_file.readline()
		head = input_file.readline().split()[1:]
	
	df = pd.read_table(f"{file_rout_max}", sep=r"\s+", comment='#', names=head)
	df = df[df.j == 100]
	df.drop(df.tail(1).index, inplace=True) # drop last 1 rows

	Tmax = np.array(df["T"])

	return Tmax


def t_cool_heat(path, N, R, Tbase, Tmax, f_d, Sigma0, gama, R_in, R_out):
	
	df_kappa = pd.read_table(f"{path}/PRmix_{N}.txt", skiprows=1, sep="\s+")
	y_interp_P = scipy.interpolate.interp1d(df_kappa["temperature[K]"], df_kappa["kappa_P[cm2/g]"]/f_d)
	y_interp_R = scipy.interpolate.interp1d(df_kappa["temperature[K]"], df_kappa["kappa_R[cm2/g]"]/f_d)

	kappaP_arr = np.empty(len(Tbase))
	kappaR_arr = np.empty(len(Tbase))

	for i, t in enumerate(Tbase):
		kappaP_arr[i] = float(y_interp_P(t))
		kappaR_arr[i] = float(y_interp_R(t))


	Sigma_gas = Sigma_g(R, Sigma0, gama, R_in, R_out)
	
	tauP = f_d * Sigma_gas * kappaP_arr
	tauR = f_d * Sigma_gas * kappaR_arr


	# xP = c_v * (1 + 2*tauP + 3/2*tauP**2) / kappaP_arr
	# xR = c_v * (1 + 2*tauR + 3/2*tauR**2) / kappaR_arr

	# time_cool_P = xP / 4 / SBc / Tmax**3
	# time_cool_R = xR / 4 / SBc / Tmax**3


	# Lacc = 450 * Lsun
	# F0 = Lacc / 4 / np.pi / (R*AU)**2

	# time_heat_P = xP * Tbase / 2 / mu / F0
	# time_heat_R = xR * Tbase / 2 / mu / F0


	# return time_cool_P, time_cool_R, time_heat_P, time_heat_R


	x = c_v * (1 + 2*tauP + 3/2*tauP*tauR) / kappaP_arr

	time_cool = x / 4 / SBc / Tmax**3


	Lacc = 450 * Lsun
	F0 = Lacc / 4 / np.pi / (R*AU)**2

	time_heat = x * Tbase / 2 / mu / F0


	return time_cool, time_heat
```
